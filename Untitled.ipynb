## Numerical Data Distribution Data

```python
def DistributionNumericalData(numerical_col):
    for col in numerical_col:
        fig = px.histogram(
                data_frame=medical_df,
                x = medical_df[col],
                marginal="box",
                nbins=47,
                title="Distribution of {}".format(col),
                )
        fig.update_layout(bargap=0.1)
        fig.show()
```

```python
DistributionNumericalData(numerical_col)
```

## n vs n analysis

```python
def ScatterNtoN(numerical_col):
    for col in numerical_col:
        for col1 in numerical_col: 
            fig = px.scatter(medical_df, 
                     x=col, 
                     y=col1,  
                     opacity=0.8,  
                     title='{} vs. {}'.format(col,col1))
            
            fig.show()
```

```python
ScatterNtoN(numerical_col)
```

## Linear Regression 

## 1.null value handle
        no null value in our dataset

## 2. one hot encoding
       category col convert numerical data

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(drop="first",dtype=int)
```

```python
ohe.fit(medical_df[["sex","smoker","region"]])
```

```python
ohe.categories_
```

```python
ohe.transform(medical_df[["sex","smoker","region"]]).toarray()
```

```python
medical_df[["s","smoker_code","northwest","southeast","southwest"]] = ohe.transform(medical_df[["sex","smoker","region"]]).toarray()
```

```python
medical_df
```

```python
medical_df.drop(["sex","smoker","region"],axis=1,inplace=True)
```

```python
medical_df
```

```python
x = medical_df.drop("charges",axis=1)
```

```python
x
```

```python
y = medical_df["charges"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lin = LinearRegression()
```

```python
lin.fit(x_train,y_train)
```

```python
lin.predict(x_test)
```

```python
lin.score(x_test,y_test)
```
